# Okay, now we know it works, let's have a list of: places (address, prices, office sizes & max occupancy)

```python
places_list = []

for page in range(1,9):
    html = requests.get(url_ny + str(page))
    soup = BeautifulSoup(html.content, "html.parser")
    name_offering = soup.find_all("p", attrs={"class":"style_title__NcI0n"})
    
    for i in range(len(name_offering)):
        name_offering[i].getText()
        places_list.append(name_offering)

    
places_list
```

```python
pricing_list = []

for page in range(1,9):
    html = requests.get(url_ny + str(page))
    soup = BeautifulSoup(html.content, "html.parser")
    price_offering = soup.find_all("p", attrs={"class":"style_pricing__TKbkD"})
    
    for i in range(len(price_offering)):
        price_offering[i].getText()
        pricing_list.append(price_offering)

    
pricing_list
```

```python
len(pricing_list)
```

```python
empty_list_price = []

for i in pricing_list:
    for item in i:
        empty_list_price.append(item)
empty_list_price 
```

```python
pricing_list2 = []

for i in empty_list_price:
    only_prices = i.getText().strip()
    if "$" in only_prices:
        pricing_list2.append(only_prices)
    else:
        pricing_list2.append("To ask")
pricing_list2[1]
```

```python
len(pricing_list2)
```

```python
pricing_list3 = []

for i in pricing_list2:
    pricing_list3.append(i[:7])
pricing_list3[:5]
```

```python
len(pricing_list3)
```

```python
pricing_list4 = []

for i in pricing_list3:
    if "$":
        pricing_list4.append(i.replace("$","").replace(",",""))
    else:
        pricing_list4.append
pricing_list4[:5]
```

```python
len(pricing_list4)
```

```python
size_list = []

for page in range(1,9):
    html = requests.get(url_ny + str(page))
    soup = BeautifulSoup(html.content, "html.parser")
    size_offering = soup.find_all("span", attrs={"class":"style_squarefeet__q7bHB"})
    
    for i in range(len(size_offering)):
        size_offering[i].getText()
        size_list.append(size_offering)

    
size_list[1]
```

```python
len(size_list)
```

```python
empty_list_size = []

for i in size_list:
    for item in i:
        empty_list_size.append(item)
empty_list_size[1]
```

```python
size_list2 = []

for i in empty_list_size:
    only_sizes = i.getText()
    size_list2.append(only_sizes[:6])
size_list2[1]
```

```python
len(size_list2)
```

```python
occupancy_list = []

for page in range(1,9):
    html = requests.get(url_ny + str(page))
    soup = BeautifulSoup(html.content, "html.parser")
    occ_offering = soup.find_all("span", attrs={"class":"style_maxOccupancy__uKsmM"})
    
    for i in range(len(occ_offering)):
        occ_offering[i].getText()
        occupancy_list.append(occ_offering)

    
occupancy_list[1]
```

```python
len(occupancy_list)
```

```python
empty_list_occ = []

for i in occupancy_list:
    for item in i:
        empty_list_occ.append(item)
empty_list_occ[1]
```

```python
occupancy_list2 = []

for i in empty_list_occ:
    only_occ = i.getText()
    occupancy_list2.append(only_occ[1:])
occupancy_list2[1]
```

```python
len(occupancy_list2)
```

# Now that we have all the info needed, we can create a pandas dataframe with all of this!

```python
nyc_rent_dict = {
    "Address" : places_list2,
    "Price" : pricing_list4,
    "Size" : size_list2,
    "Max_occupancy" : occupancy_list2
}
```

```python
nyc_real_state = pd.DataFrame(nyc_rent_dict)
nyc_real_state[:5]
```

# Getting rid of the duplicated ads

```python
nyc_real_state2 = nyc_real_state.copy()
nyc_real_state2.drop_duplicates(subset=['Address'],keep="first").reset_index().drop(["index"],axis=1)
```

```python
nyc_real_state2.to_csv("nyc_real_state_with_all_info.csv")
```

```python
nyc_real_state_clean = nyc_real_state2.copy()
nyc_real_state_clean.drop(nyc_real_state_clean[nyc_real_state_clean['Price'] == "To ask"].index, inplace = True)
nyc_real_state_clean[:6]
```

```python
nyc_real_state_clean = nyc_real_state_clean.reset_index(drop=True)
nyc_real_state_clean[:10]
```

```python
nyc_real_state_clean2 = nyc_real_state_clean.copy()
nyc_real_state_clean2["Price"] = nyc_real_state_clean2['Price'].astype(int)
type(nyc_real_state_clean2.Price[0])
```

```python
nyc_real_state_clean3 = nyc_real_state_clean2.copy()
nyc_real_state_clean4 = nyc_real_state_clean3.sort_values(by="Price", ascending=False).reset_index().drop(['index'], axis=1)
nyc_real_state_clean4[:5]
```

# Seems there were still a few duplicated values, so let's remove them again as we are only interested in those that are showing price

```python
nyc_real_state_clean5 = nyc_real_state_clean4.copy()
nyc_real_state_clean5.drop_duplicates(subset=['Address'],keep="first").reset_index().drop(["index"],axis=1)
```

```python
nyc_real_state_clean5.to_csv("nyc_real_state_only_price_info.csv")
```
