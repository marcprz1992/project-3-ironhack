```python
# First of all, let's have a look at the airports in New York
```

```python
query = "airport"
```

```python
def get_results_from_foursquare (query, location, limit=50):
       
    ll = f"{location[1]}%2C{location[0]}"
    url = f"https://api.foursquare.com/v3/places/search?query={query}&ll={ll}&limit={str(limit)}&&radius=5000"

    headers = {
        "accept": "application/json",
        "Authorization": token_fsq,
    }
    
    response = requests.get(url, headers=headers).json()
    
    return response
```

```python
nyc_location = [-73.979247,40.755716]
```
